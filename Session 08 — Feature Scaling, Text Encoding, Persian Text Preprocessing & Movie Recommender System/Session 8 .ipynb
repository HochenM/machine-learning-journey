```python
#onehot is for words - countvectorize for text
#Example in text
from sklearn.feature_extraction.text import CountVectorizer

docs = ['cat dog cat', 'dog bird']

vectorizer = CountVectorizer()
result = vectorizer.fit_transform(docs)

#          bird  cat  dog
# Doc 1:  [0,   2,   1]   ← 2 cats, 1 dog, 0 birds
# Doc 2:  [1,   0,   1]   ← 1 bird, 0 cats, 1 dog
result.toarray()
```

```python
vec2 = CountVectorizer()
vec2.fit_transform([docs[0]])
```

CountVectorizer: "cat" appears 5 times → score = 5

TF-IDF:          "cat" appears 5 times, BUT "cat" appears in EVERY document
                  → score = LOW (common word = not important)

                 "python" appears 2 times, BUT only in 1 document
                  → score = HIGH (rare word = very important!)


                  TF-IDF = TF × IDF

TF (Term Frequency)        = How often word appears in THIS document
IDF (Inverse Document Freq) = How RARE the word is across ALL documents

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
vectorizer.fit_transform(name).toarray()
print(vectorize.get_feature_names_out())
vectorize.vocabulary_
```

```python
docs = ['cat dog cat', 'dog bird']

vectorizer = TfidfVectorizer()
result = vectorizer.fit_transform(docs)

result.toarray()

```

```python
CountVectorizer:                    TF-IDF:
          bird  cat  dog                  bird   cat   dog
Doc 1:  [  0,    2,   1 ]         Doc 1: [0.00, 0.88, 0.38]
Doc 2:  [  1,    0,   1 ]         Doc 2: [0.65, 0.00, 0.65]

CountVectorizer:                   TF-IDF:
  "cat"=2 (raw count)               "cat"=0.88 (high importance —
                                   appears twice but only in 1 doc!)
  "dog"=1 in both docs              "dog"=0.38 (low importance —
                                   appears in BOTH docs — common word)
```

### Text Pre_processing 

```python
from parsivar import Tokenizer, Normalizer
tokenizer = Tokenizer()
normalizer= Normalizer()
```

```python
punctuation_list= "!@#$%^,،؟?،"
import re
def pre_processing(text):
    text = normalizer.normalize(text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = "".join([char for char in text if char not in punctuation_list])
    tokens = tokenizer.tokenize_words(text)
    return text, tokens

pre_processing('سلاااام، چه طورییییی؟')
```

```python
#ُsentences tokenize
text = 'سلاااام، چه طورییییی؟.من که عالیم مرسییی'
tokens = tokenizer.tokenize_sentences(text)

text = ([char for char in text if char not in punctuation_list])
tokens
"".join(text)
```

```python
# Stemming
from parsivar import FindStems

stemmer = FindStems()
print(stemmer.convert_to_stem('کتابی'))
print(stemmer.convert_to_stem('کتاب های'))
print(stemmer.convert_to_stem('کتاب '))

print(stemmer.convert_to_stem('میخوانم'))
print(stemmer.convert_to_stem('میخوانیم'))

```
