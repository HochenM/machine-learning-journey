```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([100,150,170,200,220,250,300,350,400]).reshape(-1,1)    # متراژ
y = np.array([ 20, 22, 24, 31, 33, 34, 42, 44, 49])
```

```python
plt.scatter(X, y)
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Tehran House Price")
```

```python
cost_function = []
```

```python
# 1 iter
a = 0
b = 0

y_pred = a * X + b

y_pred

plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Tehran House Price")

plt.scatter(X, y, c="b", label="Real Data")
plt.plot(X,y_pred, c="r", label="Predict")
```

```python
error = y - y_pred
square_error = np.square(error)
mean_square_error = np.mean(square_error)
root_mean_square_error = np.sqrt(mean_square_error)
print("RMSE :", root_mean_square_error)

cost_function.append(root_mean_square_error)
```

```python
# 2 iter (solver ==> a=0.1, b=0.5)
a = 0.1
b = 0.5

y_pred = a * X + b

y_pred

plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Tehran House Price")

plt.scatter(X, y, c="b", label="Real Data")
plt.plot(X,y_pred, c="r", label="Predict")
```

```python
error = y - y_pred
square_error = np.square(error)
mean_square_error = np.mean(square_error)
root_mean_square_error = np.sqrt(mean_square_error)
print("RMSE :", root_mean_square_error)

cost_function.append(root_mean_square_error)
```

```python
plt.plot(cost_function, "r", label="Cost Function")
```

```python
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Tehran House Price")

plt.scatter(X, y, c="b", label="Real Data")
cost_function = []
parameters = []

for a in np.arange(-1,1.5, 0.1):
    for b in np.arange(0, 60, 5):
        y_pred = a * X + b
        
        plt.plot(X,y_pred, c="r", label="Predict")
        
        error = y - y_pred
        square_error = np.square(error)
        mean_square_error = np.mean(square_error)
        root_mean_square_error = np.sqrt(mean_square_error)
        #print("RMSE :", root_mean_square_error)
        
        cost_function.append(root_mean_square_error)
        parameters.append([a,b])
```

```python
plt.plot(cost_function, "r", label="Cost Function")
```

```python
result = min(cost_function)
result_idx = cost_function.index(result)
result_idx
```
