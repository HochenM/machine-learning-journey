```python
# Linear Regression Model + Gradient Descent
# Train loop
# Verbose
X = np.array([100,200,300,400,500]).reshape(-1,1)
y = np.array([ 20, 22, 29, 30, 35]).reshape(-1,1)
X
n_samples, n_feature = X.shape
X_bias = np.c_[np.ones((n_samples, 1)), X]

theta = np.array([[0], [0]])


learning_rate = 0.00001
max_iter = 25

cost_function = []

for i in range(max_iter):
    y_pred = X_bias @ theta

    error = y_pred - y

    # Solver (Gradient Descent)
    gradient = X_bias.T @ error / n_samples
    theta = theta - gradient * learning_rate

    # Metric
    mse = np.mean(np.square(y - y_pred))
    cost_function.append(mse)
    print(f"Iter : {i+1} - Loss : {mse}")
    
```

```python
plt.plot(cost_function, "r")
```

```python
X.shape
y.shape
n_samples, n_feature = X.shape
X_bias = np.c_[np.ones((n_samples,1)),X]
X_bias.T
```

```python
# Hyper parametes
# we changes this to tune the best model we can find
learning_rate = 0.000004
max_iter = 200
thetas = np.array([1,20])

```

```python
cost_function = []

for i in range(max_iter):
    y_pred = X_bias @ thetas

    
    error  = y_pred - y

    gradient = X_bias.T @ error / n_samples

    thetas = thetas - gradient * learning_rate
    mse = np.mean(np.square(error))
    rmse = np.sqrt(mse)
    cost_function.append(rmse)
    print(f"Iter : {i+1} - Loss : {mse}") #verbose

    
```
